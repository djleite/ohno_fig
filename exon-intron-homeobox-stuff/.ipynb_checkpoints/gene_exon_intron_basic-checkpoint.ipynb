```python
import sys
import re
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# clean the gtf, collecting the transcript and the feature of choice
def clean_gff(fle, string):
    out_list = []
    with open(fle, 'r') as inf:
        for line in inf:
            l = line.split("\t")
            if l[2] == "transcript" and l[8].endswith(".t1\n") or l[2] == string and """.t1";""" in l[8]:
                out_list.append(line)
    return out_list

# break down the gff to get each gene
def get_section(fle):
    current_section = []
    for line in fle:
        if line.lstrip().split("\t")[2] == "transcript" and line.endswith("t1\n"):
            if current_section:
                yield current_section
            current_section = [line]
        else:
            current_section.append(line)
    if current_section:
        yield current_section

# process the last column to get the gene name
def process(instr):
    instr = instr.replace('"', '').replace(';','').split(' ')
    out = instr[1]
    return out

# add the gff information to dictionary
def add2dict(line, gene_n, gene_dict, count, i):
    if process(line[8]) == gene_n:
        if int(count) == int(len(i))-1:
            count += 1
            gene_dict[gene_n][str(line[2]+"_"+str(count))] = int(int(line[4])-int(line[3]))
            count = "last"
            gene_dict[gene_n][str(line[2]+"_"+str(count))] = int(int(line[4])-int(line[3]))
        else:
            count += 1
            gene_dict[gene_n][str(line[2]+"_"+str(count))] = int(int(line[4])-int(line[3]))
        return count, gene_dict

# input clean gff list and turn to dictionary of dictionaries for the length of each feature { gene1 {f1:len,f2:len}, gene2 {f1:len,f2:len} }
def sections2dict(sections):
    gene_dict = {}

    # for each chunk which contains a transcript and feature ie intron
    for i in sections:
        #check if positive or negative strand and get other features
        x = i[0].strip().split("\t")
        if x[2] == "transcript":
            gene_dict[x[8]] = {}
            #name of gene
            gene_n = x[8]
            #start the intron count
            count = 0
            #strand
            strand = x[6]
            #pop off first line
            i.pop(0)
        # for + and - strand count feature and also get last
        if strand == '+':
            for line in i:
                line = line.strip().split("\t")
                count, gene_dict = add2dict(line, gene_n, gene_dict, count, i)                
        elif strand == '-':
            for line in reversed(i):
                line = line.strip().split("\t")
                count, gene_dict = add2dict(line, gene_n, gene_dict, count, i)

    return gene_dict

# plot the dictionary
def dict2plot(gene_dict, feature):
    df = pd.DataFrame(gene_dict)
    ndf = df.filter(items=[feature+"_1",
                           feature+"_2",
                           feature+"_3",
                           feature+"_4",
                           feature+"_5",
                           feature+"_6",
                           feature+"_7",
                           feature+"_8",
                           feature+"_9",
                           feature+"_10",
                           feature+"_11",
                           feature+"_12",
                           feature+"_13",
                           feature+"_14",
                           feature+"_15",
                           feature+"_16",
                           feature+"_17",
                           feature+"_18",
                           feature+"_19",
                           feature+"_20",
                           feature+"_last"
                          ], axis = 0)

    # Plotting
    plt.figure(figsize=(5, 5))
    sns.boxplot(data=ndf.T, orient='h', showfliers=False)
    #plt.xscale('log')
    plt.show()
    
    return ndf, gene_dict
    
def wrapper(infile, feature, name):
    intron_list = clean_gff(infile, feature)
    intron_section = list(get_section(intron_list))
    gene_dict = sections2dict(intron_section)
    ndf,gene_dict = dict2plot(gene_dict, feature)
    ndf.to_csv(name+"_"+feature+".csv")
    return ndf,gene_dict
```

```python
library(reshape2)
library(ggplot2)

read_files_and_make_lists <- function(file1, file2, file3, file4, file5) {
    # Read terms from files
    terms1 <- paste0(readLines(file1), ".t1")
    terms2 <- paste0(readLines(file2), ".t1")
    terms3 <- paste0(readLines(file3), ".t1")
    terms4 <- paste0(readLines(file4), ".t1")
    terms5 <- paste0(readLines(file5), ".t1")
    
    # Remove the first 7 characters from each term
    terms1 <- substr(terms1, 8, nchar(terms1))
    terms2 <- substr(terms2, 8, nchar(terms2))
    terms3 <- substr(terms3, 8, nchar(terms3))
    terms4 <- substr(terms4, 8, nchar(terms4))
    terms5 <- substr(terms5, 8, nchar(terms5))
    
    
    # Return a list of the terms
    return(list(terms1, terms2, terms3, terms4, terms5))
}


# Function to add column to the data frame
add_column_to_data <- function(data, file1, file2, file3, file4, file5) {
    # Read terms and make lists
    lists <- read_files_and_make_lists(file1, file2, file3, file4, file5)
    
    # Match terms to Var2 and add as a new column with list names
    for (i in 1:length(lists)) {
        match_idx <- match(lists[[i]], data$Var2)
        new_col_name <- paste0("list", i)
        data[new_col_name] <- NA
        data[new_col_name][!is.na(match_idx)] <- new_col_name
    }
    
    # Return the modified data frame
    return(data)
}


data <- as.matrix(read.csv("CARROT.csv", header=T, row.names=1))
melted_data <- melt(data)
# Example usage:
file1 <- "/work/WGD/GO/from_braker/ohno_list/CARROT_1.5_0.txt"
file2 <- "/work/WGD/GO/from_braker/ohno_list/CARROT_1.5_B.txt"
file3 <- "/work/WGD/GO/from_braker/ohno_list/CARROT_1.5_C.txt"
file4 <- "/work/WGD/GO/from_braker/ohno_list/CARROT_1e-10_segmental.txt"
file5 <- "/work/WGD/GO/from_braker/ohno_list/CARROT_1e-10_tandem.txt"

lists <- read_files_and_make_lists(file1, file2, file3, file4, file5)

# Add group column based on group keys
melted_data$group <- ifelse(melted_data$Var2 %in% unlist(lists[1]), "ohno",
                     ifelse(melted_data$Var2 %in% unlist(lists[2]), "seg",
                     ifelse(melted_data$Var2 %in% unlist(lists[3]), "tan",
                     ifelse(melted_data$Var2 %in% unlist(lists[4]), "seg2",
                     ifelse(melted_data$Var2 %in% unlist(lists[5]), "tan2",'no_ann')))))

# Change outlier, color, shape and size
ggplot(melted_data, aes(x=value, y=Var1,fill=group)) + 
    geom_boxplot(outlier.shape = NA)+
    scale_x_continuous(limits = c(0, 20000))
```
